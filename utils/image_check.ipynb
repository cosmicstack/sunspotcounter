```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('../data/test/20230102_110000_512_HMIIF.jpg')
```

```python
img
```

```python
img_array = np.array(img)
img_array.shape
```

```python
def plot_matrix(matrix):
    plt.figure(figsize=(10, 10))
    plt.matshow(matrix)
    plt.colorbar()
    plt.show()
```

```python
plot_matrix(img_array)
```

```python
img_array_mask = (img_array > 0) & (img_array < 90)
```

```python
plot_matrix(img_array_mask)
```
